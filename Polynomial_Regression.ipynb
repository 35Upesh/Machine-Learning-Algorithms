```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Load Dataset
df_sal = pd.read_csv('Position_Salaries.csv')
df_sal.head()
```

```python
# Describe Dataset
df_sal.describe()
```

```python
# Data Distribution
plt.title("Salary Distribution plot")
sns.distplot(df_sal['Salary'])
plt.show()
```

```python
plt.scatter(df_sal['Level'],df_sal['Salary'], color='Lightcoral')
plt.title('Salary vs Level')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Split dataset into dependent and independent
X = df_sal.iloc[:,1:-1]
y = df_sal.iloc[:,-1]
```

```python
# Train the model
lr = LinearRegression()
lr.fit(X,y)

pr = PolynomialFeatures(degree=4)
X_poly = pr.fit_transform(X)
lr2 = LinearRegression()
lr2.fit(X_poly,y)
```

```python
# Predict results
y_pred = lr.predict(X)
y_pred2 = lr2.predict(X_poly)
```

```python
# Visualize the result

# With Linear Regression
plt.scatter(X,y, color='Lightcoral')
plt.plot(X,y_pred, color='firebrick')
plt.title('Salary vs Level Linear (Real Data)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend(['X/y', 'X/y_pred'], title='salary/Level', loc='best', facecolor='white')
plt.show()
```

```python
#with Polynomial Regression
plt.scatter(X,y, color='Lightcoral')
plt.plot(X,y_pred2, color='firebrick')
plt.title('Salary vs Level Polynomial (Real Data)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend(['X/y', 'X/y_pred2'], title='salary/Level', loc='best', facecolor='white')
plt.show()
```

```python
# Let's Predict some values
print(f'Linear Regression result : {lr.predict([[7.5]])}')
print(f'Polynomial Regression result : {lr2.predict(pr.fit_transform([[7.5]]))}')
```
